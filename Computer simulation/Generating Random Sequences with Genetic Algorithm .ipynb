```python
from numpy import zeros, count_nonzero
from numpy.random import rand, randint
from random import choice
from seaborn import heatmap
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt 
```

```python
# Initialization parameter
HIGHT = 10
WIDTH = 10

color_mat = zeros((HIGHT,WIDTH))
color_mat
```

```python
def change_start(color_mat, Hight=8, Width=8):
    zero_index_list = []
    for i in range(0, Hight):
        for j in range(0, Width):
            if(color_mat[i][j]==0):
                zero_index_list.append((i,j))
    
    print("state changed")
    arr = choice(zero_index_list)
    x = arr[0]
    y = arr[1]
    return (x, y)
```

```python
x, y = change_start(color_mat, HIGHT, WIDTH)
print("start x is", x, "and start y is", y)
```

```python
# R : 1
# G : 2
# B : 3
i = 0

while 0. in color_mat.flatten():
    
    print(f"__________{i}__________")
    
    if (x>=0 and y>=0 and y<HIGHT and x<HIGHT):
        rand_num = rand(1)
        
    
        # state function
        if(color_mat[x][y]==0):
            print('State painted')
            print(f"({x}, {y})")
            if(rand_num>=0 and rand_num<=(1/3)):
                color_mat[x][y]=1
            if(rand_num>(1/3) and rand_num<=(2/3)):
                color_mat[x][y]=2
            if(rand_num>(2/3) and rand_num<=1):
                color_mat[x][y]=3

        # Up function
        elif(rand_num>=0 and rand_num<=(1/4)):
            y+=1
            print('Up')
            print(f"({x}, {y})")
            if(y>=HIGHT):
                x, y=change_start(color_mat, HIGHT, WIDTH)
            elif(color_mat[x][y]==0):
                print('painted')
                if(rand_num>=0 and rand_num<=(1/3)):
                    color_mat[x][y]=1
                if(rand_num>(1/3) and rand_num<=(2/3)):
                    color_mat[x][y]=2
                if(rand_num>(2/3) and rand_num<=1):
                    color_mat[x][y]=3

        # Down function
        elif(rand_num>(1/4) and rand_num<=(2/4)):
            y-=1
            print("Down")
            print(f"({x}, {y})")
            if(y<0):
                x, y=change_start(color_mat, HIGHT, WIDTH)
            elif(color_mat[x][y]==0):
                print("painted")
                if(rand_num>=0 and rand_num<=(1/3)):
                    color_mat[x][y]=1
                if(rand_num>(1/3) and rand_num<=(2/3)):
                    color_mat[x][y]=2
                if(rand_num>(2/3) and rand_num<=1):
                    color_mat[x][y]=3


        # Right function
        elif(rand_num>(2/4) and rand_num<=(3/4)):
            x+=1
            print("Right")
            print(f"({x}, {y})")
            if(x>=HIGHT):
                x, y=change_start(color_mat, HIGHT, WIDTH)
            elif(color_mat[x][y]==0):
                print("painted")
                if(rand_num>=0 and rand_num<=(1/3)):
                    color_mat[x][y]=1
                if(rand_num>(1/3) and rand_num<=(2/3)):
                    color_mat[x][y]=2
                if(rand_num>(2/3) and rand_num<=1):
                    color_mat[x][y]=3

        # Left function
        elif(rand_num>(3/4) and rand_num<=1):
            x-=1
            print("Left")
            print(f"({x}, {y})")
            if(x<0):
                x, y=change_start(color_mat, HIGHT, WIDTH)
            elif(color_mat[x][y]==0):
                print("painted")
                if(rand_num>=0 and rand_num<=(1/3)):
                    color_mat[x][y]=1
                if(rand_num>(1/3) and rand_num<=(2/3)):
                    color_mat[x][y]=2
                if(rand_num>(2/3) and rand_num<=1):
                    color_mat[x][y]=3

                           
    else:
        x, y=change_start(color_mat, HIGHT, WIDTH)
        
    print()
    
    i+=1
```

```python
red_number = count_nonzero(color_mat == 1)
green_number = count_nonzero(color_mat == 2)
blue_number = count_nonzero(color_mat == 3)
```

```python
print(f"Red :{red_number}, Green :{green_number}, Blue :{blue_number}")
heatmap(data=color_mat, cbar=True, xticklabels=False, yticklabels=False,
        cmap=ListedColormap(['red', 'green', 'blue']),linewidths=1.5, 
        linecolor='white')

plt.show()
```
