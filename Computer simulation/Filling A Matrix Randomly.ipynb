```python
from numpy import array, sort
from random import choice
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# parameters
x = 2546
arr = array([9678, 4758, 3415, 9975])
```

# Functions

```python
# median_cross function
def median_cross(input_arr, x):
            
    # choose medain 4 number
    for index in range(len(input_arr)): 
        ans = input_arr[index]*x
        input_arr[index] = int(str(ans)[2:6])
        
    # insurense that number length always be 4   
    for index in range(len(input_arr)):
        if input_arr[index]<1000:
            while True:
                input_arr[index] = input_arr[index]*10
                if(input_arr[index]>9999):
                    input_arr[index] = input_arr[index]/10
                    break
       
    return input_arr

# cross_over
def cross_over(number1, number2):
    part1 = str(number1)[0:2]
    part2 = str(number2)[2:4]
    ans = int(part1+part2)
    return ans

# fitness function
def fitness_function(input_arr):
    arr = sort(input_arr)
    arr[1] = cross_over(arr[1], arr[3])
    arr[2] = cross_over(arr[2], arr[0])
    return arr
```

# Main

```python
# main function
ans_list = []
for i in range(500):
    arr = median_cross(arr, x)
    arr = fitness_function(arr)
    ans = choice(arr)
    ans_list.append(ans)
```

# Visualization

```python
ans_list = array(ans_list)

plt.figure(figsize=(14,5))
plt.plot(ans_list[:200],'b--o')
plt.show()
```
